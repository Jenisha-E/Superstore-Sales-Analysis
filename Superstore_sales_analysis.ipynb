Converts Order Date and Ship Date into datetime objects for easier time-based analysis.

```python
# Convert 'Order Date' and 'Ship Date' to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'], dayfirst=True)
df['Ship Date'] = pd.to_datetime(df['Ship Date'], dayfirst=True)

# Preview changes
df[['Order Date', 'Ship Date']].head()

```

Matplotlib is for basic plotting, and Seaborn makes plots more attractive and easier to interpret.

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: make plots look nicer
sns.set_style('whitegrid')
%matplotlib inline
```

Shows total sales per product category. Helps identify which categories contribute most to revenue.

```python
plt.figure(figsize=(8,5))
sns.barplot(data=df, x='Category', y='Sales', estimator=sum, ci=None, palette='viridis')
plt.title('Total Sales by Category')
plt.xlabel('Category')
plt.ylabel('Sales ($)')
plt.show()
```

Breaks down sales by sub-category, helping understand which products are top sellers.

```python
plt.figure(figsize=(12,6))
sns.barplot(data=df, x='Sub-Category', y='Sales', estimator=sum, ci=None, palette='magma')
plt.title('Total Sales by Sub-Category')
plt.xticks(rotation=45)
plt.xlabel('Sub-Category')
plt.ylabel('Sales ($)')
plt.show()
```

Compares sales across regions to find the best-performing areas.

```python
plt.figure(figsize=(7,5))
sns.barplot(data=df, x='Region', y='Sales', estimator=sum, ci=None, palette='pastel')
plt.title('Sales by Region')
plt.ylabel('Sales ($)')
plt.show()
```

Shows how sales fluctuate across months, useful for spotting seasonal trends.

```python
# Aggregate monthly sales
monthly_sales = df.groupby('Month')['Sales'].sum().reindex([
    'January','February','March','April','May','June','July','August',
    'September','October','November','December'
])

plt.figure(figsize=(12,6))
monthly_sales.plot(kind='line', marker='o', color='teal')
plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('Sales ($)')
plt.grid(True)
plt.show()
```

Identifies the most valuable customers who contribute the most to revenue.

```python
top_customers = df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(x=top_customers.values, y=top_customers.index, palette='cubehelix')
plt.title('Top 10 Customers by Total Sales')
plt.xlabel('Sales ($)')
plt.show()
```

Shows the distribution of shipping delays, highlighting how efficiently orders are delivered.

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Ship_Delay'], kde=True, color='#FFB6C1', bins=20)  # Light pink
plt.title('Shipping Delay Distribution', fontsize=16)
plt.xlabel('Shipping Delay (days)')
plt.show()

```

Shows relationships between numeric variables like Sales, Quantity, and Ship_Delay. Useful for detecting patterns and dependencies.

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
